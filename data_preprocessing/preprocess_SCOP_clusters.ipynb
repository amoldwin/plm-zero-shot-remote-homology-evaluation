```python
# generate dataset at different threshold from sampling cluster representative.


def get_cluster_dict_from_cdhitclusters(th=90):
    # if the cluster dict is already computed for usage, return that
    # clus_dict_path = f"data/tmp/scop_sf_seq_cluster_dict_at_{th}_seq_identity.pkl"
    # if osp.exists(clus_dict_path):
    #     cluster_dict = Utils.load_pickle(clus_dict_path)
    #     print(f"    num of clusters: {len(cluster_dict)}") #18575 at 40% identity
    #     return cluster_dict
    
    #else
    print("\tComputing cluster dictionary.")
    
    f = open(home_dir + f"data/SCOP/clusters_fa_represeq/at_th_{th}.clstr", mode="r")
    flag=False
    cluster_dict = dict()
    cluster_no = 0
    for i, line in enumerate(f.readlines()):
        # print(i, line)
        
        if ">Cluster" not in line:
            line_items = line.split(" ")
            cluster.append(str(line_items[1][1:-3]))
            # print(cluster)

        elif ">Cluster" in line:
            if flag: 
                # out.writelines(" ".join(cluster)) #if want to see the clusters, comment out this line
                # out.write("\n")
                
                cluster_dict[f"c{cluster_no}"] = cluster
                cluster_no += 1
            
            cluster = []
            flag = True
            # print(" ".join(cluster))
            
    print(f"\tnum of clusters: {len(cluster_dict)}") #18575 at 40% identity
    # Utils.save_as_pickle(cluster_dict, clus_dict_path)
    # print(cluster_dict)
    return cluster_dict
```

```python
for th in [10, 20, 30, 40, 70, 95]:
    # th=30
    th = str(th)

    clstr_dict = get_cluster_dict_from_cdhitclusters(th)
    # print(clstr_dict["c0"]) # printing first cluster

    samples = []
    for clstr_key, clstr_values in clstr_dict.items():
        # print(clstr_key, clstr_values)
        samples.append(random.choice(clstr_values))
        # break

    # samples


    out_dir = home_dir + f"data/SCOP/processed_at_th/"
    out_df = scop_df[scop_df["FA-DOMID"].apply(str).isin(samples)]
    print(th, out_df.shape)
    out_df.to_csv(f"{out_dir}th_{th}.tsv", sep="\t", header=True, index=False)
    # break
```
